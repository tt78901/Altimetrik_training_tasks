```python
import pandas as pd
import numpy as np
data={
    "Name":["Alice","Bob","Charlie"],
    "Age":[25,30,35],
    "City":["New York",np.nan,"Los Angeles"],
    "Salary":[10000,200000,np.nan],
    "Department":["HR",np.nan,"IT"]
}
```

```python
df=pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
df=df.fillna({'City':'City Missing','Salary':0.0,})
```

```python
df
```

```python
df.dropna(axis=1)
```

```python
df.fillna({"Salary" : df["Salary"].mean()},inplace=True)
df
```

```python
sales={
    "Date" :["2024-01-01","2024-01-02","2024-01-03","2024-01-04"],
    "Region":["East","West",np.nan,"south"],
    "Units_Sold":[10,np.nan,15,12],
    "Units_Price":[300,250,400,np.nan]
}
```

```python
dff=pd.DataFrame(sales)
dff
```

```python
dff.isnull().sum()
```

```python
dff.fillna({"Region":"Region Missing","Units_Price":dff["Units_Price"].mean()},inplace=True)
dff
```

```python
dff
```

```python
dff.groupby("Region")["Units_Sold"].mean()
```

```python
dff.groupby("Region").agg({"Units_Sold":"sum","Units_Price":"mean"})
```
