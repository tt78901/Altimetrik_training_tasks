```python
Q1 = df1['total_bill'].quantile(0.25)

Q3 = df1['total_bill'].quantile(0.75)

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df1= df1[(df1['total_bill'] >= lower_bound) & (df1['total_bill'] <= upper_bound)]

# for tip
Q1 = df1['tip'].quantile(0.25)
Q3 = df1['tip'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df1= df1[(df1['tip'] >= lower_bound) & (df1['tip'] <= upper_bound)]

# for size

Q1 = df1['size'].quantile(0.25)
Q3 = df1['size'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df1= df1[(df1['size'] >= lower_bound) & (df1['size'] <= upper_bound)]
```
